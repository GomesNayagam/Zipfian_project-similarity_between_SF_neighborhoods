# Building a similarity matrix

I am now going to make a similarity matrix so that I can see how similar the different hoods are based on quantitative measures.  I measured multiple types of crime in past year, quality of life issues in past year, types of business open/opening in the past year, fog, types of houses for sale (e.g., # of bedrooms), tranportation connectivity to the rest of the city.  

```python
import pandas as pd
import numpy as np
import csv
from sklearn.metrics import pairwise
from sklearn.decomposition import PCA
%pylab inline
```

```python
#open the data
sf_hoods=pd.read_csv(open('../hoods_data/final_to_analyze/SF_hoods_density_full_1-27-14.csv'))
del sf_hoods['Unnamed: 0']

feature_names = filter(lambda x: x not in ['hood'], sf_hoods.columns)

#data without the hood name; all float data that can be run through algorithms
X = sf_hoods[feature_names]
#the names of the hoods
hood_names = sf_hoods['hood']
```

## Run PCA on data set to reduce dataset

Many of my features are highly correlated.  For example, I have multiple types of 311 (quality of life) calls and I'm sure many of them reflect a similar latent factor.  If I didn't reduce these data, I would bias the similarity matrix to this latent factor, simply because there are more 'quality of life' features (e.g., more columns describing types of crime) than features reflecting other types of important latent factors.   

I am going to plot variance explained by each component to decide how many to keep.  I want to keep basically all of the variance explained by my variables (99%); I just want to reduce redundant variables. 

```python
#total number of columns/features in data set
X.shape
```

```python
#--MANUALLY CHANGEABLE--

#amount of variance I want to explain
explained_variance = .99

#number of components to show on the plot
components = 50

# -- * --

#see how much variance is explained if I reduce it down
pca = PCA(n_components = components).fit(X)

#A function that will tell you how many components to retain if you want to explain a certain portion of variance (0-1)?
def retain(portion):
    for i,v in enumerate(pca.explained_variance_ratio_.cumsum()):
        if v > portion:
            return i + 1
            break

figure(figsize = (20, 7))
plt.bar(arange(components) + .5, pca.explained_variance_ratio_.cumsum())
plt.ylim((0, 1))

#Run the retain function here; 

components = retain(explained_variance)
pylab.title("keep " + str(components) + ' components to explain ' + str(explained_variance) + ' of the variance', fontsize=30)


#other axis labels
plt.xlabel('No. of principal components', fontsize=20)
plt.ylabel('Cumulative variance explained', fontsize=20)
plt.grid(axis = 'y', ls = '-', lw = 1, color = 'black')
```

I want to keep almost all of information that I have on the hoods.  I just want to reduce redundancy so that redundant features do not get overcounted (redundancy is a byproduct of the way I got the information (e.g., crime having 30 features)) .  So I am going to keep the number of components that explain 99% of the origianl dataset.  This reduces the number of variables from 98 to 42.  

```python
print 'You need to keep ' + str(components) + ' components' + ' to explain ' + str(explained_variance) + ' of your variance.'
pca = PCA(n_components = components).fit(X)
hoods_reduced = pca.fit_transform(X)
```

```python
pd.DataFrame(hoods_reduced).to_csv('../hoods_data/final_to_analyze/hoods_reducedfeatures.csv')
```

## Compute similarity matrix

```python
#set the type of similarity metric here (e.g., euclidean, correlation, cosine)
similarity_metric = 'correlation'
#create similarity matrix
hoods_similarity_np = pairwise.pairwise_distances(hoods_reduced, metric=similarity_metric, n_jobs=4)

#make a csv copy
hoods_similarity = pd.DataFrame(hoods_similarity_np)
hoods_similarity.to_csv('../hoods_data/final_to_analyze/hoods_similarity_matrix.csv')
```

## EXAMINE THE MATRIX!   

I am going to examine the matrix in a few ways, in part to see if similarity matrix does a good job assessing similarities between hoods.  I also want to see the matrix produces any interesting insights about hoods in SF.  

First make lookup tables.

```python
hood_rev_lookup = dict(hood_names)
#hood_rev_lookup

#create dictionary that will allow me to find the row number of hood/district
hood_lookup={}
for k,v in hood_rev_lookup.items():
    hood_lookup[v]=k
```

Write a function that is going to return you a list of similar hoods.  1st argument should be target hood, 2cd argument should be number of similar hoods to list, move from most similar to less similar

```python
#find similar hoods 
def similar_hoods(hood, numb_hoods):
    #select row of hood you are interested in, returning list of similarity scores / how similar each hood is to target hood
    hoodlist=hoods_similarity.ix[hood_lookup[hood],:]
    
    #now use hood_rev_lookup to get names of most similar hoods
    similar_hoods=[]
    for k, v in hood_rev_lookup.items():
        for i in range(1,numb_hoods+1):
            simhood=pd.Series(np.sort(hoodlist)).iloc[i:i+1].index.values
            if k == simhood:
                similar_hoods.append(str(i)+'. '+v)
    return 'Sim to ' + str(hood) + ': ' + str(sorted(similar_hoods))
```

test it out on hoods that I know to see if it seem reasonable (it does!)

```python
#These are hoods I know about and can assess whether the top 5 similar hoods are reasonable
print similar_hoods('Haight Ashbury', 5)
print similar_hoods('Outer Sunset', 5)
print similar_hoods('Outer Richmond', 5)
print similar_hoods('Russian Hill', 5)
print similar_hoods('Noe Valley', 5)
print similar_hoods('Castro', 5)
print similar_hoods('Mission', 5)
print similar_hoods('Potrero Hill', 5)
print similar_hoods('Tenderloin', 5)
print similar_hoods('Dogpatch', 5)
print similar_hoods('Excelsior', 5)
print similar_hoods('Visitacion Valley', 5)
print similar_hoods('Hunters Point', 5)
print similar_hoods('Bayview', 5)
print similar_hoods('Marina', 5)
#they seem highly reasonable!
```

```python
#RUN THIS TO FIND TO find top three similar hoods for all neighborhoods in SF
for i in hood_names:
    print similar_hoods(i, 4)
```

Given what I know (my domain knowledge) about SF, the similarity matrix appears to do a fantantastic job capturing hood similarity.   The results are similar to what I would give. 

Interestingly, even though geographic information (e.g., closeness) was not a feature, often areas close to each other are labeled as being similar.  

# Hierarchical cluster of the data

I want to visualize which neighborhoods are similar to each other. 

```python
# perform clustering and plot the dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram
titles = np.array(hood_names)

#make the dendogram using the hoods_similarity matrix
figure(figsize(10,30))
R = dendrogram(linkage(hoods_similarity, method='complete'), labels = titles, leaf_font_size=10, orientation='left' )
    #use labels above to pass a vector that contains the values you want to have as the labels of the datapoints
xlabel('Hoods', fontsize=4)
ylabel('Height')
suptitle('SF Hoods Cluster Dendrogram', fontweight='bold', fontsize=14);
```

Very cool!  The difference between red and green neighborhoods is noticeable.  For the most part, red hoods are busy and/or rich, while green hoods are sleepier and/or poorer.  

The mission is paired up with SoMa (particularly the more sketchy part of it; i.e., not South Beach - the area near AT&T park).  This pairing is similar to other busy areas with lots of restaurants and bars (e.g., tenderloin, north beach, castro, polk gulch, etc.)


# Multidimensional scaling

Another visual representation of neighborhood similarities.  Again, my features seem to capture similaries between SF hoods rather well.  There is a sub cluster that seem to represent parts of town that are very dead (e.g. TI, India basin, etc.)

```python
from sklearn.manifold import MDS

#performce MDS
mds=MDS(dissimilarity='precomputed')
hood_mds=mds.fit_transform(hoods_similarity)
hood_names = np.array(hood_names)

#Plot the MDS
figsize(25, 15)
plt.scatter(hood_mds[:, 0], hood_mds[:, 1],  marker='+', alpha=0)
plt.title('Similarity between Hoods in San Francisco')

for i in hoods_similarity:
    plt.annotate(hood_names[i], (hood_mds[i, 0], hood_mds[i, 1]),
                 color = 'red',
                 alpha = 0.7, 
                 fontsize = 12,
                 horizontalalignment = 'center', 
                 verticalalignment = 'center',
                 family = 'sans-serif')
# Turn off axes labels.
plt.setp(plt.gca().get_yaxis(), visible = False)
plt.setp(plt.gca().get_xaxis(), visible = False)
plt.show()
```
